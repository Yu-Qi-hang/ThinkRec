```python
import json
import numpy as np
from matplotlib import pyplot as plt
import os
```

```python
file_path = '/home/yuqihang/projects/CoLLM/collm-datasets/booknew/id2title.json'
```

```python
id2title = json.load(open(file_path, 'r'))
```

```python
des = [meta['description'] for meta in id2title.values()]
```

```python
from collections import defaultdict
lengths = defaultdict(int)
for item in des:
    lengths[len(item)] += 1
```

```python
#对lengths按照键排序
sorted_lengths = sorted(lengths.items(), key=lambda x: x[0], reverse=False)
sorted_lengths = {k: v for k, v in sorted_lengths}
lengths = sorted_lengths
```

```python
for i in sorted_lengths:
    if sorted_lengths[i] >50:
        lengths[i] = 50
        print(i,sorted_lengths[i])
```

```python
list(lengths.keys())[-10:]
```

```python
import pandas as pd
import seaborn as sns

# 将 lengths 转换为 DataFrame
lengths_df = pd.DataFrame(list(lengths.items()), columns=['Length', 'Count'])
# 使用 seaborn 绘制柱状图
plt.figure(figsize=(10, 6))
sns.barplot(x='Length', y='Count', data=lengths_df)
plt.title('Distribution of Lengths')
plt.show()
```
